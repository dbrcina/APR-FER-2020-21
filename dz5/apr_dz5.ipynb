### 5. Domaća zadaća -  _Algoritmi numeričke integracije_ ###

Učitavanje potrebnih biblioteka:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%pylab inline
```

Pomoćne funkcije:

```python
def analytical_func(x1_0, x2_0, t_array):
    x1_array = x1_0 * np.cos(t_array) + x2_0 * np.sin(t_array)
    x2_array = x2_0 * np.cos(t_array) - x1_0 * np.sin(t_array)
    return x1_array, x2_array

def cummulative_error(x1_real, x1_analytical, x2_real, x2_analytical):
    return np.sum(np.absolute(x1_real - x1_analytical)) + np.sum(np.absolute(x2_real - x2_analytical))

def make_plots(x1_0, x2_0, t_array, x1_real, x2_real, x1_analytical=None, x2_analytical=None, legend=''):
    plt.figure(figsize=(15, 5))
    for i in range(1, 3):
        ax = plt.subplot(1, 2, i)
        ax.grid(True)
        if x1_analytical is not None and x2_analytical is not None:
            ax.plot(t_array, x1_analytical if i==1 else x2_analytical, label='analytical')
        ax.plot(t_array, x1_real if i==1 else x2_real, label=legend)
        ax.set_xlabel('t')
        ax.set_ylabel(f'x{i}')
    plt.legend()
    plt.show()

def do_task(task_num, alg, x1_0=None, x2_0=None, do_analytics=False):
    df = pd.read_csv(f'data/task{task_num}/{alg}.csv')
    t_array = df['t']
    x1_real, x2_real = df['x1'], df['x2']
    x1_analytical, x2_analytical = None, None
    if do_analytics:
        x1_analytical, x2_analytical = analytical_func(x1_0, x2_0, t_array)
        print(f'Kumulativna pogreška iznosi: {cummulative_error(x1_real, x1_analytical, x2_real, x2_analytical)}')
    make_plots(x1_0, x2_0, t_array, x1_real, x2_real, x1_analytical, x2_analytical, legend=alg)
```

### 1. Zadatak ###

Početne varijable stanja:

```python
x1_0 = 1
x2_0 = 1
```

#### Euler ####

```python
do_task(1, 'euler', x1_0, x2_0, do_analytics=True)
```

```python
do_task(1, 'reversed_euler', x1_0, x2_0, do_analytics=True)
```

```python
do_task(1, 'pece2', x1_0, x2_0, do_analytics=True)
```
